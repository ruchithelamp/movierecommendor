```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
```

```python
G = nx.karate_club_graph()

```

```python
nx.draw(G, with_labels=True)
```

```python
# this is a social network pertaining to a Karate club that has been stuied by Wayne Zachary
# we are looking at a network of 34 members of a club, and their interactions outside of the club
```

```python
# centrality 
deg_cent = nx.degree_centrality(G)
bet_cent = nx.betweenness_centrality(G)

# what are the biggest centralities
deg_sort = sorted(deg_cent.items(), key=lambda x: x[1], reverse=True)
bet_sort = sorted(bet_cent.items(), key=lambda x: x[1], reverse=True)
```

```python
# print out the top three centralities

print(deg_sort[:3])
print(bet_sort[:3])

```

```python
# community detection 
communities = community.greedy_modularity_communities(G)

```

```python
print(f"Detecting {len(communities)} communities")

for i, c in enumerate(communities):
    print(f"community {i}: {list(c)}")
```

```python
colors = []
for n in G.nodes():
    for i, c in enumerate(communities):
        if n in c:
            colors.append(i)

plt.figure(figsize=(5, 4))
```
